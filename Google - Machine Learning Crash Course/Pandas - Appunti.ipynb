# Appunti Pandas

##Overview del package
Pandas è un package Python che fornisce delle strutture di dati per lavorare con dati "labellati" in modo facile ed intuitivo. Viene usato per il data analysis e la manipolazione dei dati. 

Le due strutture dati principali di pandas sono:
* ***Series*** - 1-Dimensionale. Rappresenta un array labellato, monodimensionale ed omogeneo.
* ***DataFrame*** - 2 - Dimensionale. Struttura tabulare generalmente bidimensionale e labellata che può cambiare dimensione e con delle colonne potenzialmente etereogenee.

Pandas è costruito al di sopra di *NumPy* e si integra bene con libreire di terze parti.

Di più su Pandas al seguente indirizzo:
https://pandas.pydata.org/docs/getting_started/overview.html
##DataFrame di Pandas

Questo Colab introduce i [**DataFrames**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html), che costituisce la struttura dati centrale dell'APU Pandas.

Un DataFrame è simile ad un foglio di calcolo (tipo Excel) tenuto in memoria. Come un foglio di calcolo:
* Un Dataframe memorizza i dati in celle
* Un Dataframe ha (di solito) delle colonne provviste di nomi e delle righe numerate. 

##Importare i moduli NumPy e Pandas


```
import numpy as np
import pandas as pd
```

## Creare un DataFrame
Il seguente codice crea un semplice DataFrame contenente 10 celle organizzate nel seguente modo:
* 5 righe
* 2 colonne, una chiamate `temperatura` ed un'altra chiamata `attività`

Il codice segente istanzia una classe `pd.DataFrame` per generare un DataFrame. La classe prende in input due argomenti:
* Il primo argomento fornisce i dati per popolare le 10 celle. Questo frammento di codice chiama `np.array` per generare una matrice NumPy 5X2.
* Il secondo argomento identifica i nomi delle due colonne


```
# Crea e popola un NumPy array da 5x2.
my_data = np.array([[0, 3], [10, 7], [20, 9], [30, 14], [40, 15]])

# Crea una lista Pyton che mantiene i nomi delle due colonne.
my_column_names = ['temperature', 'activity']

# Crea un DataFrame.
my_dataframe = pd.DataFrame(data=my_data, columns=my_column_names)

# DataFrame Print dell'intero Dataframe
print(my_dataframe)
```

## Aggiungere una nuova colonna al DataFrame



Possiamo aggiungere una nuova colonna ad un DataFrame semplicemente assegnando dei valori ad una colonna con un nome nuovo. Per esempio, il codice seguente crea una terza colonna chiamata `adjusted` all'interno del DataFrame `my_dataframe`:

```
# Crea una nuova colonna chiamata adjusted
my_dataframe["adjusted"] = my_dataframe["activity"] + 2

# Print dell'intero DataFrame
print(my_dataframe)
```

## Specificare un sottoinsieme del DataFrame

Pandas fornisce modi diversi per isolare delle righe, colonne o porzioni di celle in un Dataframe.

```
print("Riga #0, #1, and #2:")
print(my_dataframe.head(3), '\n')

print("Riga #2:")
print(my_dataframe.iloc[[2]], '\n')

print("Riga #1, #2, and #3:")
print(my_dataframe[1:4], '\n')

print("Colonna 'temperature':")
print(my_dataframe['temperature'])
```

## Copying a DataFrame (optional)
## Copiare un DataFrame (opzionale)
Pandas fornisce due modi differenti per duplicare un DataFrame:

* **Riferimento.** Se assegni un DataFrame ad una nuova variabili, ogni cambiamento al DataFrame o alla nuova variabile saranno riflessi l'uno nell'altro.
* **Copiandolo.** Se chiamiamo il metodo `pd.DataFrame.copy`, creiamo una vera copia indipendente. I cambiamenti al DataFrame originale o alla copia non saranno riflessi l'uno nell'altro.

Questa differenza è sottile, ma importante.
